```python
def train_one_epoch_kulum(model, optimizer, data_loader, device, epoch, print_freq, scaler=None):
    model.train()
    metric_logger = utils.MetricLogger(delimiter="  ")
    metric_logger.add_meter("lr", utils.SmoothedValue(window_size=1, fmt="{value:.6f}"))
    header = f"Epoch: [{epoch}]"

    lr_scheduler = None
    if epoch == 0:
        warmup_factor = 1.0 / 1000
        warmup_iters = min(1000, len(data_loader) - 1)

        lr_scheduler = torch.optim.lr_scheduler.LinearLR(
            optimizer, start_factor=warmup_factor, total_iters=warmup_iters
        )

    for images, targets in metric_logger.log_every(data_loader, print_freq, header):
        images = list(image.to(device) for image in images)
        targets = [{k: v.to(device) for k, v in t.items()} for t in targets]
        with torch.cuda.amp.autocast(enabled=scaler is not None):
            loss_dict = model(images, targets)
            losses = sum(loss for loss in loss_dict.values())

        # reduce losses over all GPUs for logging purposes
        loss_dict_reduced = utils.reduce_dict(loss_dict)
        losses_reduced = sum(loss for loss in loss_dict_reduced.values())

        loss_value = losses_reduced.item()

        if not math.isfinite(loss_value):
            print(f"Loss is {loss_value}, stopping training")
            print(loss_dict_reduced)
            sys.exit(1)

        optimizer.zero_grad()
        if scaler is not None:
            scaler.scale(losses).backward()
            scaler.step(optimizer)
            scaler.update()
        else:
            losses.backward()
            optimizer.step()

        if lr_scheduler is not None:
            lr_scheduler.step()

        metric_logger.update(loss=losses_reduced, **loss_dict_reduced)
        metric_logger.update(lr=optimizer.param_groups[0]["lr"])

    return metric_logger
```

```python
import torch
import numpy as np
import monai.data
import monai.losses
import monai.metrics
import torch.utils.data
import logging
import traceback
import torch.nn as nn

from tqdm import tqdm
from datetime import datetime
from .early_stopping import EarlyStopping
from typing import List, Union
from monai.utils import first
from models.model import ModelUNet
from monai.data.utils import decollate_batch
from monai.transforms import EnsureType, Compose, AsDiscrete
from torch.optim.swa_utils import AveragedModel, SWALR
from torch.optim.lr_scheduler import CosineAnnealingWarmRestarts
from torch.utils.tensorboard import SummaryWriter

class Trainer:
    """Executes a typical Pytorch training loop
    
    Args:
        model: the model used for training
            Required: True
        train_loader (torch.utils.data.DataLoader): DataLoader obj for the training dataset
            Required: True
        val_loader (torch.utils.data.DataLoader): DataLoader obj for the validation dataset   
            Required: True
        optimizer (torch.optim): Optimization scheme used for the training process
            Required: True
        loss (monai.losses): Criterion to be minimized during training
            Required: True
        metric (monai.metrics): Metric to be used for measuring model efficiency
            Required: True
        learning_rate (float): Learning rate to be used for updating model parameters during Gradient Descent
            Required: True
        early_stopping (EarlyStopping): EarlyStopping obj to stop training process if no significant improve is being made - to prevent overfitting
            Required: False     Default: None, i.e. do not use Early Stopping
        epochs (int): number of epochs to train
            Required: False     Default: 100
        device (string): device on which the training process is running i.e. either cpu or cuda
            Required: False     Default: cpu - because cuda might not be available
        val_interval (int): every how many epochs to compute validation error
            Required: False     Default: 2
        dry_run (bool): perform a dryn_run i.e. a pass though the model without kicking off the training process
            Required: False     Default: False
        retrain (bool): boolean indicating whether it is a fresh training or attempt to fine-tune existing model
            Required: False     Default: False
        load_path (str): path indicating where the pretrained model is located
            Required: False     Default: None
    """
    def __init__(self, 
                 model, 
                 train_loader: torch.utils.data.DataLoader, 
                 val_loader: torch.utils.data.DataLoader,  
                 optimizer: torch.optim, 
                 loss: monai.losses,
                 metric: monai.metrics,
                 save_path: str,
                 learning_rate: float=1e-4,
                 early_stopping: EarlyStopping=None, 
                 epochs: int=100,
                 device: str='cpu', 
                 val_interval: int=2,
                 dry_run: bool=False,
                 retrain: bool=False,
                 load_path: str=None):
        self.retrain = retrain
        self.device = device
        if self.retrain:
            self.model = self.load_checkpoint(load_path=load_path)
        else:
            self.model = model
        self.model = nn.DataParallel(self.model)
        self.model.to(device=self.device) # move model to device before initializing optimizer
        self.train_loader = train_loader
        self.val_loader = val_loader
        self.epochs = epochs
        self.optimizer = optimizer(self.model.parameters(), lr=learning_rate, weight_decay=1e-5)
        self.loss = loss
        self.metric = metric
        self.metric_val = metric
        self.val_interval = val_interval
        self.save_path = save_path
        self.early_stopping = early_stopping
        self.dry_run = dry_run
        self.writer = SummaryWriter()
        self._logger = logging.getLogger(__name__)
    
    def fit(self):
        """Typical PyTorch Training Loop

        Args:
            none
        Returns:
            output_dict (dict): Dictionary containing model, optimizer, loss, training_loss, validation_loss and validation metric
        
        """
        scheduler = CosineAnnealingWarmRestarts(optimizer=self.optimizer, T_0=30, eta_min=1e-4)
        val_outputs = []
        val_labels = []
        post_pred = Compose([EnsureType(), AsDiscrete(argmax=True, to_onehot=3, num_classes=3)])
        post_label = Compose([EnsureType()])
        self._logger.info(f'Attemping training with     \n' \
                          f'Optimizer: {self.optimizer} \n' \
                          f'Loss: {self.loss}           \n' \
                          f'Metric: {self.metric}       \n' \
                          f'Scheduler: {scheduler}      \n' )

        if self.dry_run:
            self._logger.info('Running a DRY RUN, will pass through the model once.')
            sample = first(self.train_loader)
            output = self.model(sample['image'].float().to(self.device))
            return {}
        train_losses = []
        all_train_losses = []
        all_val_losses = []
        metrics = []
        for epoch in range(1, self.epochs + 1):
            self._logger.info(f'Epoch {epoch} / {self.epochs}')
            self.model.train()
            for batch, train_data in enumerate(self.train_loader, 1):
                
                image, segmentation = train_data['image'], train_data['segmentation']
                image, segmentation = image.float().to(self.device), segmentation.float().to(self.device)
                try:
                
                    # zero gradients
                    self.optimizer.zero_grad()
                    # compute predictions
                    output = self.model(image)
                    
                    if (epoch == 50 or epoch == 100 or epoch == 150 or epoch == 200  or epoch == 300) and batch == 20:
                        image_write = image[0,0,:,:,1].clone().detach()  # save 2nd slice only i.e. right kidney
                        segmentation_write = segmentation[0,0,:,:,1].clone().detach()
                        output_write = output[0,0,:,:,1].clone().detach()
                        output_write = torch.softmax(output_write, dim=1)
                        # write to .tfevents
                        self.writer.add_image(tag=f'Train_CT{epoch}_{batch}_{datetime.now()})', img_tensor=image_write, dataformats='HW')
                        self.writer.add_image(tag=f'Train_ground{epoch}_{batch}_{datetime.now()}', img_tensor=segmentation_write, dataformats='HW')
                        self.writer.add_image(tag=f'Train_pred{epoch}_{batch}_{datetime.now()}', img_tensor=output_write, dataformats='HW')
                    
                    # compute loss
                    loss = self.loss(output, segmentation)
                    val_outputs = [post_pred(i) for i in decollate_batch(output)]
                    val_labels = [post_label(i) for i in decollate_batch(segmentation)]
                    
                    # compute metric for current iteration
                    self.metric(y_pred=val_outputs, y=val_labels)
                    
                    # perform backprop
                    loss.backward()
                    self.optimizer.step()
                    
                except Exception as e:
                    self._logger.info(f'Exception caught while training in {traceback.format_exc()}. Aborting...')
                    return {'exception': e}
                
                # keep track of train loss
                train_losses.append(loss.item())
            
            metric = self.metric.aggregate().item()
            self.metric.reset()
            scheduler.step()
            
            if epoch % self.val_interval == 0: # validate every val_interval epochs
                self.save_checkpoint(model_state_dict=self.model.state_dict(), optimizer_state_dict=self.optimizer.state_dict(), loss=None,
                                    train_loss=None,
                                    val_loss=None,
                                    val_metric=None, 
                                    save_path=self.save_path.joinpath(f'trained_model_epoch{epoch}.tar'))

            all_train_losses.append(np.average(train_losses))
            self.writer.add_scalar(tag='Loss/train', scalar_value=all_train_losses[-1], global_step=epoch)
            if epoch % self.val_interval == 0:
                # validate model and get average loss
                val_loss, val_metric = self.validation_loop()
                self.writer.add_scalar(tag='Loss/eval', scalar_value=val_loss, global_step=epoch)
                self.writer.add_scalar(tag='Metric/eval', scalar_value=val_metric, global_step=epoch)
                all_val_losses.append(val_loss)
                metrics.append(val_metric)
                self._logger.info(f'Validation loss: {all_val_losses[-1]:.12f} \n' \
                                  f'Metric: {metrics[-1]:.12f} \n')
                
                # if self.early_stopping is not None:
                #     self.early_stopping(all_val_losses[-1], self.model)
                #     if self.early_stopping.early_stop:
                #         self._logger.info("Early Stoppping")
                #         break  
            self._logger.info(f'Finished epoch: {epoch} of {self.epochs} with ' +
                              f'train loss: {all_train_losses[-1]:.12f} ' + 
                              f'train metric: {metric:.12f}')
            train_losses = []   
        # load the last checkpoint with the best model
        #self.model.load_state_dict(torch.load(self.early_stopping.path))
        output_dict = {'model_state_dict': self.model, 'optimizer_state_dict': self.optimizer, 'loss': self.loss, # maybe self.optimizer.state_dict
                       'train_loss': all_train_losses, 'val_loss': all_val_losses, 'val_metric': metrics}
        return output_dict

    def validation_loop(self):
        """Validation loss computation Loop

        Args:
            none
        Returns:
            np.average(val_losses), metric (List): Average loss during validation and validation metric
        
        """
        val_losses = [] # batch loss
        val_outputs = []
        val_labels = []
        post_pred = Compose([EnsureType(), AsDiscrete(argmax=True, to_onehot=3, num_classes=3)])
        post_label = Compose([EnsureType()])#, AsDiscrete(to_onehot=3, num_classes=3)])
        with torch.no_grad():
            # evaluate model
            self.model.eval()
            for val_batch, val_data in enumerate(self.val_loader, 1):
                #self._logger.info(f'Validating batch {val_batch}')
                val_image, val_segm = val_data['image'].float().to(self.device), val_data['segmentation'].float().to(self.device)
                try:
                    # compute predictions
                    val_output = self.model(val_image)
                    # compute loss
                    loss = self.loss(val_output, val_segm)
                    val_outputs = [post_pred(i) for i in decollate_batch(val_output)]
                    val_labels = [post_label(i) for i in decollate_batch(val_segm)]
                    # compute metric for current iteration
                    self.metric_val(y_pred=val_outputs, y=val_labels)
                except Exception as e:
                    self._logger.info(f'Exception caught while validating in {traceback.format_exc()}. Aborting...')
                    return np.inf
                # record loss
                val_losses.append(loss.item())
            metric = self.metric_val.aggregate().item()
            self.metric_val.reset()
        return np.average(val_losses), metric

    def save_checkpoint(self, 
                        model_state_dict, 
                        optimizer_state_dict, 
                        loss, 
                        train_loss: List, 
                        val_loss: List, 
                        val_metric: List, 
                        save_path=None):
        """Save checkpoint while training the model

        Args:
            model_state_dict (dict): Dictionary containing model state i.e. weights and biases
                Required: True
            optimizer_state_dict (dict): Dictionary containing optimizer state i.e. gradients
                Required: True
            loss (monai.losses): Loss criterion state data
                Required: True
            train_loss (List): List containing training losses 
                Required: True
            val_loss (List): DList containing validation losses 
                Required: True
            save_path (str): Path to save the checkpoint
                Required: False     Default: None  
        Returns:
            -
        """
        torch.save({'model_state_dict': model_state_dict,
                    'optimizer_state_dict': optimizer_state_dict,
                    # 'loss': loss,
                    # 'train_loss': train_loss,
                    # 'val_loss': val_loss,
                    # 'val_metric': val_metric,
                    # 'test_metric': -np.inf # not yet evaluated on test set
                    }, save_path)

    def load_checkpoint(self, load_path):
        """Load pretrained model (state dictionary) - To be used for fine-tuning

        Args:
            load_path (str): path indicating where the pretrained model is located
                Required: True  
        Returns:
            model (model.ModelUNET): Model obj once loaded with state dictionary of pretrained model   
        """
        if load_path is not None:
            checkpoint = torch.load(load_path)
        model = ModelUNet(inputs=1, outputs=3)
        model_dict = checkpoint['model_state_dict']
        new_dict = {}
        for k,v in model_dict.items():
            new_dict[k[7:]] = v
        model.load_state_dict(new_dict)
        return model
        # model.load_state_dict(checkpoint['model_state_dict'])
        # return model
```
